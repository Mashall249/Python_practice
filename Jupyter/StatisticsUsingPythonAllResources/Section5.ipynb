```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/Piece_Dim.csv')
df.head()
```

```python
plt.hist(df['Length'])
plt.show
```

```python
plt.hist(df['Length'],bins=15)
plt.show()
```

```python
plt.hist(df['Length'],bins=15, color='red')
plt.xlabel('Length in mm')
plt.ylabel('Count')
plt.title('Histgram of piece length')
plt.show()
```

```python
plt.boxplot(df['Length'])
plt.show()
```

```python
op1 = df[df['Operator'] == 'Op-1']
op2 = df[df['Operator'] == 'Op-2']
```

```python
op1.describe()
```

```python
op2.describe()
```

```python
plt.boxplot([op1['Length'], op2['Length']])
plt.show()
```

```python
plt.boxplot([op1['Length'], op2['Length']], notch=True)
plt.xlabel('Operator')
plt.ylabel('Length in mm')
plt.title('Box Plot by Operator')
plt.show()
```

```python
plt.plot(df['Length'])
mean_l = df['Length'].mean()
std_l = df['Length'].std()
plt.axhline(y= mean_l,color='red',linestyle='-')
plt.axhline(y= mean_l + 3 * std_l,color='red',linestyle='--')
plt.show()
```

```python
plt.scatter(x=df['Length'], y=df['Width'],marker='v')
plt.show()
```
