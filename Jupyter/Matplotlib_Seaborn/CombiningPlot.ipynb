```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(100)
sns.histplot(dataset,bins=25, kde=True)
plt.show()
```

```python
sns.histplot(dataset, bins=25, kde=True, element='poly', color='orange')
plt.show()
```

```python
# KDEだけ制御したい場合（ラベル・色）
sns.histplot(dataset, bins=25, color='lightblue', label='Histogram')
sns.kdeplot(dataset, color='indianred', linewidth=2, label='KDE Plot')
sns.rugplot(dataset, color='black')
plt.legend()
plt.show()
```

```python
ser1 = Series(dataset,name='My Data')
```

```python
sns.histplot(ser1, bins=25, kde=True)
plt.show()
```
