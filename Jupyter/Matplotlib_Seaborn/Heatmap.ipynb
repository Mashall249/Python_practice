```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flight_dframe = flights.pivot(index='month', columns='year', values='passengers')
flight_dframe
```

```python
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_dframe, annot=True, fmt='d')
```

```python
center_value = flight_dframe.loc["Jan", 1955]
sns.heatmap(flight_dframe, center=center_value)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# 
yearly_totals = flight_dframe.sum()
year_dframe = pd.DataFrame({
    'year': yearly_totals.index,
    'flights': yearly_totals.values
})

sns.barplot(x='year', y='flights', data=year_dframe, ax=ax1)

# 
sns.heatmap(
    flight_dframe,
    cmap="Blues",
    cbar_kws={'orientation': 'horizontal'},
    ax=ax2
)
```

```python
sns.clustermap(flight_dframe)
```

```python
sns.clustermap(flight_dframe, col_cluster=False)
```

```python
sns.clustermap(flight_dframe, standard_scale=1)
```
