```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flight_dframe = flights.pivot(index='month', columns='year', values='passengers')
flight_dframe
```

```python
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_dframe, annot=True, fmt='d')
```

```python
center_value = flight_dframe.loc["Jan", 1955]
sns.heatmap(flight_dframe, center=center_value)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# 年ごとの合計便数を棒グラフで可視化
yearly_totals = flight_dframe.sum()
year_dframe = pd.DataFrame({
    'year': yearly_totals.index,
    'flights': yearly_totals.values
})

sns.barplot(x='year', y='flights', data=year_dframe, ax=ax1)

# 下にヒートマップを描画（カラーバーを横向きに）
sns.heatmap(
    flight_dframe,
    cmap="Blues",
    cbar_kws={'orientation': 'horizontal'},
    ax=ax2
)
```
