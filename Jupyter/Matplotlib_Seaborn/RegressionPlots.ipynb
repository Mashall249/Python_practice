```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(
  x='total_bill', y='tip', data=tips,
  scatter_kws={'marker': 'o', 'color': 'indianred'},
  line_kws={'linewidth': 1, 'color': 'blue'}
  )
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, order=4)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
sns.lmplot(x='size', y='tip_pct', data=tips)
```

```python
sns.lmplot(x='size', y='tip_pct', data=tips, x_jitter=0.1)
```

```python
sns.lmplot(
    x='size', y='tip_pct', data=tips,
    x_estimator=np.mean  # 
)
```

```python
sns.lmplot(
    x='total_bill', y='tip_pct', data=tips,
    hue='sex', markers=['x', 'o']
)
```

```python
sns.lmplot(x='total_bill', y='tip_pct', data=tips, hue='day')
```

```python
sns.lmplot(
    x='total_bill', y='tip_pct', data=tips,
    lowess=True,
    line_kws={'color': 'black'}
)
```

```python
sns.regplot(x='total_bill', y='tip_pct', data=tips)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

sns.regplot(x='total_bill', y='tip_pct', data=tips, ax=ax1)
sns.violinplot(x='size', y='tip_pct', data=tips, palette='Reds_r', ax=ax2)
```
