```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
plt.show()
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
plt.show()
```

```python
sns.boxplot(data1,orient="h")
plt.show()
```

```python
# データ1：正規分布
data1 = stats.norm(loc=0, scale=5).rvs(100)

# データ2：左右対称なガンマ分布を結合
gamma1 = stats.gamma(5).rvs(50) - 1
gamma2 = -1 * stats.gamma(5).rvs(50)
data2 = np.concatenate([gamma1, gamma2])

# Box Plot：違いが分かりにくい
sns.boxplot(data=[data1, data2], whis=np.inf)
plt.title("Box Plot (May Miss Distribution Shape)")
plt.show()
```

```python
sns.violinplot(data=[data1,data2])
plt.show()
```

```python
sns.violinplot(data=[data1, data2], bw=0.1)
plt.show()
```

```python
sns.violinplot(data1,inner='stick')
plt.show()
```
