```python
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
plt.show()
```

```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred')
plt.show()
```

```python
plt.hist(dataset1, density=True, color='indianred', alpha=0.5, bins=20)
plt.hist(dataset2, density=True, alpha=0.5, bins=20)
plt.show()
```
